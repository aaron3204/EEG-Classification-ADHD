```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import mne
from sklearn.model_selection import GroupKFold
from sklearn.metrics import (classification_report, roc_auc_score, 
                             confusion_matrix, roc_curve, auc)
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler, LabelEncoder
import tensorflow as tf
from tensorflow.keras import layers, models, utils, callbacks, regularizers
from mne.preprocessing import ICA
from scipy.stats import kurtosis
from scipy.signal import welch

# Configure plots
plt.style.use('ggplot')
sns.set_palette("husl")

# ---------- Parameters ----------
CSV_PATH = "adhdata.csv"
SFREQ = 128.0         # Hz
EPOCH_LENGTH = 2.0    # seconds
TEST_SIZE = 0.2       # fraction of subjects for testing
BATCH_SIZE = 32
EPOCHS = 100
DROP_RATE = 0.5       # dropout rate
L2_REG = 1e-4         # weight decay coefficient

# Frequency bands for feature extraction
BANDS = {
    'delta': (1, 4),
    'theta': (4, 8),
    'alpha': (8, 13),
    'beta': (13, 30),
    'gamma': (30, 40)
}

# ------------------------
# EEGNet Architecture
# ------------------------
def EEGNet(nb_classes=2, Chans=19, Samples=256,
           dropoutRate=DROP_RATE, kernLength=64, F1=8, D=2, F2=16, l2=L2_REG):
    input1 = layers.Input((Chans, Samples, 1))
    
    x = layers.Conv2D(F1, (1, kernLength), padding='same', use_bias=False,
                      kernel_regularizer=regularizers.l2(l2))(input1)
    x = layers.BatchNormalization()(x)
    
    x = layers.DepthwiseConv2D((Chans, 1), use_bias=False, depth_multiplier=D,
                              depthwise_regularizer=regularizers.l2(l2))(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('elu')(x)
    x = layers.AveragePooling2D((1, 4))(x)
    x = layers.Dropout(dropoutRate)(x)
    
    x = layers.SeparableConv2D(F2, (1, 16), use_bias=False, padding='same',
                              depthwise_regularizer=regularizers.l2(l2),
                              pointwise_regularizer=regularizers.l2(l2))(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('elu')(x)
    x = layers.AveragePooling2D((1, 8))(x)
    x = layers.Dropout(dropoutRate)(x)
    
    x = layers.Flatten()(x)
    x = layers.Dense(nb_classes, kernel_regularizer=regularizers.l2(l2))(x)
    output = layers.Activation('softmax')(x)
    
    return models.Model(inputs=input1, outputs=output)

# ------------------------
# DeepConvNet Architecture
# ------------------------
def DeepConvNet(nb_classes=2, Chans=19, Samples=256, dropoutRate=DROP_RATE, l2=L2_REG):
    input1 = layers.Input((Chans, Samples, 1))
    
    x = layers.Conv2D(25, (1, 5), kernel_regularizer=regularizers.l2(l2))(input1)
    x = layers.Conv2D(25, (Chans, 1), kernel_regularizer=regularizers.l2(l2))(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('elu')(x)
    x = layers.MaxPooling2D((1, 2))(x)
    x = layers.Dropout(dropoutRate)(x)
    
    x = layers.Conv2D(50, (1, 5), kernel_regularizer=regularizers.l2(l2))(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('elu')(x)
    x = layers.MaxPooling2D((1, 2))(x)
    x = layers.Dropout(dropoutRate)(x)
    
    x = layers.Conv2D(100, (1, 5), kernel_regularizer=regularizers.l2(l2))(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('elu')(x)
    x = layers.MaxPooling2D((1, 2))(x)
    x = layers.Dropout(dropoutRate)(x)
    
    x = layers.Conv2D(200, (1, 5), kernel_regularizer=regularizers.l2(l2))(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('elu')(x)
    x = layers.MaxPooling2D((1, 2))(x)
    x = layers.Dropout(dropoutRate)(x)
    
    x = layers.Flatten()(x)
    x = layers.Dense(nb_classes, kernel_regularizer=regularizers.l2(l2))(x)
    output = layers.Activation('softmax')(x)
    
    return models.Model(inputs=input1, outputs=output)

# ------------------------
# Feature Extraction Function
# ------------------------
def compute_band_powers(X, sfreq, bands):
    n_epochs, n_channels, n_times = X.shape
    f, psd = welch(X.reshape(n_epochs * n_channels, n_times), fs=sfreq, nperseg=128, axis=-1)
    psd = psd.reshape(n_epochs, n_channels, -1)
    
    band_powers = []
    for band_name, (low, high) in bands.items():
        idx_band = np.where((f >= low) & (f < high))[0]
        if len(idx_band) == 0:
            continue
        power = psd[:, :, idx_band].sum(axis=2)
        band_powers.append(power)
    
    band_powers = np.stack(band_powers, axis=2)
    band_powers = band_powers.reshape(n_epochs, -1)
    return band_powers

# ------------------------
# Data Loading & Preprocessing
# ------------------------
def load_and_preprocess():
    df = pd.read_csv(CSV_PATH)
    channels = df.columns[:19].tolist()
    
    X, y, groups = [], [], []
    for pid, group in df.groupby('ID'):
        try:
            data = group[channels].T.values * 1e-6
            info = mne.create_info(channels, sfreq=SFREQ, ch_types='eeg')
            raw = mne.io.RawArray(data, info)
            
            raw.set_eeg_reference('average')
            raw.filter(1, 40, fir_design='firwin')
            
            def implement_ica(raw):
                original_raw = raw.copy()
                ica = ICA(random_state=97, max_iter=800)
                ica.fit(raw)
                sources = ica.get_sources(raw).get_data()
                component_kurtosis = kurtosis(sources, axis=1)
                kurtosis_95th_percentile = np.percentile(component_kurtosis, 95)
                components_to_exclude = np.where(component_kurtosis >= kurtosis_95th_percentile)[0]
                ica.exclude = components_to_exclude
                try:
                    cleaned_raw = ica.apply(raw)
                    return cleaned_raw
                except:
                    return original_raw
            
            raw = implement_ica(raw)
            
            events = mne.make_fixed_length_events(raw, duration=EPOCH_LENGTH)
            epochs = mne.Epochs(raw, events, tmin=0, tmax=EPOCH_LENGTH,
                               baseline=None, preload=True)
            
            if len(epochs) == 0:
                print(f"No epochs created for patient {pid}")
                continue
                
            n_epochs = len(epochs)
            X.append(epochs.get_data())
            y.extend([group['Class'].iloc[0]] * n_epochs)
            groups.extend([pid] * n_epochs)
            
        except Exception as e:
            print(f"Error processing patient {pid}: {str(e)}")
            continue

    X = np.concatenate(X)
    y = LabelEncoder().fit_transform(y)
    return X, y, groups

X, y, groups = load_and_preprocess()
print(f"Data shape: {X.shape}")
print(f"Class distribution: {np.unique(y, return_counts=True)}")

# ------------------------
# Visualization Functions
# ------------------------
def plot_metrics(history):
    fig, ax = plt.subplots(1, 2, figsize=(15, 5))
    ax[0].plot(history.history['accuracy'], label='Train')
    ax[0].plot(history.history['val_accuracy'], label='Validation')
    ax[0].set_title('Accuracy')
    ax[0].legend()
    
    ax[1].plot(history.history['loss'], label='Train')
    ax[1].plot(history.history['val_loss'], label='Validation')
    ax[1].set_title('Loss')
    # plt.savefig(f'metrics_fold.png')
    # plt.close()
    plt.show()

def plot_confusion_matrix(y_true, y_pred, classes):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6,6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=classes, yticklabels=classes)
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title('Confusion Matrix')
    # plt.savefig('confusion_matrix.png')
    # plt.close()
    plt.show()

def plot_roc(y_true, y_probs, model_name):
    fpr, tpr, _ = roc_curve(y_true, y_probs)
    roc_auc = auc(fpr, tpr)
    
    plt.figure()
    plt.plot(fpr, tpr, color='darkorange', lw=2, 
             label=f'ROC (AUC = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'ROC Curve - {model_name}')
    plt.legend(loc="lower right")
    # plt.savefig(f'roc_curve_{model_name}.png')
    # plt.close()
    plt.show()

# ------------------------
# Model Training & Evaluation
# ------------------------
def evaluate_models(X, y, groups):
    gkf = GroupKFold(n_splits=5)
    results = {'eegnet': [], 'deepconvnet': [], 'random_forest': [], 'svm': [], 'logistic_regression': []}
    
    for fold, (train_idx, test_idx) in enumerate(gkf.split(X, y, groups)):
        print(f"\n{'='*40}\nFold {fold+1}\n{'='*40}")
        
        X_train, X_test = X[train_idx], X[test_idx]
        y_train, y_test = y[train_idx], y[test_idx]
        
        # ------------------------
        # EEGNet & DeepConvNet Processing
        # ------------------------
        scaler = StandardScaler()
        n_tr, n_ch, n_t = X_train.shape
        X_train = scaler.fit_transform(X_train.reshape(n_tr, -1)).reshape(n_tr, n_ch, n_t)
        X_test = scaler.transform(X_test.reshape(len(test_idx), -1)).reshape(len(test_idx), n_ch, n_t)
        
        X_train_eeg = X_train[..., np.newaxis]
        X_test_eeg = X_test[..., np.newaxis]
        y_train_cat = utils.to_categorical(y_train)
        
        # ------------------------          
        # EEGNet Training
        # ------------------------
        print("\nTraining EEGNet...")
        eegnet_model = EEGNet(nb_classes=2, Chans=n_ch, Samples=n_t)
        eegnet_model.compile(optimizer='adam', 
                            loss='categorical_crossentropy', 
                            metrics=['accuracy'])
        
        history_eegnet = eegnet_model.fit(
            X_train_eeg, y_train_cat,
            validation_data=(X_test_eeg, utils.to_categorical(y_test)),
            epochs=EPOCHS,
            batch_size=BATCH_SIZE,
            callbacks=[
                callbacks.EarlyStopping(patience=10, restore_best_weights=True),
                callbacks.ReduceLROnPlateau(factor=0.5, patience=5)
            ],
            verbose=1
        )
        
        y_probs_eegnet = eegnet_model.predict(X_test_eeg)[:, 1]
        results['eegnet'].append(roc_auc_score(y_test, y_probs_eegnet))
        plot_metrics(history_eegnet)
        
        # ------------------------          
        # DeepConvNet Training
        # ------------------------
        print("\nTraining DeepConvNet...")
        deepconv_model = DeepConvNet(nb_classes=2, Chans=n_ch, Samples=n_t)
        deepconv_model.compile(optimizer='adam', 
                              loss='categorical_crossentropy', 
                              metrics=['accuracy'])
        
        history_deepconv = deepconv_model.fit(
            X_train_eeg, y_train_cat,
            validation_data=(X_test_eeg, utils.to_categorical(y_test)),
            epochs=EPOCHS,
            batch_size=BATCH_SIZE,
            callbacks=[
                callbacks.EarlyStopping(patience=10, restore_best_weights=True),
                callbacks.ReduceLROnPlateau(factor=0.5, patience=5)
            ],
            verbose=1
        )
        
        y_probs_deepconv = deepconv_model.predict(X_test_eeg)[:, 1]
        results['deepconvnet'].append(roc_auc_score(y_test, y_probs_deepconv))
        plot_metrics(history_deepconv)
        
        # ------------------------
        # Traditional ML Models
        # ------------------------
        ml_models = {
            'Random Forest': RandomForestClassifier(n_estimators=200),
            'SVM': make_pipeline(StandardScaler(), SVC(kernel='rbf', probability=True)),
            'Logistic Regression': make_pipeline(StandardScaler(), LogisticRegression())
        }
        
        X_train_ml = compute_band_powers(X_train, SFREQ, BANDS)
        X_test_ml = compute_band_powers(X_test, SFREQ, BANDS)
        
        for name, estimator in ml_models.items():
            print(f"\nTraining {name}...")
            estimator.fit(X_train_ml, y_train)
            y_pred = estimator.predict_proba(X_test_ml)[:, 1]
            key_name = name.lower().replace(' ', '_')
            results[key_name].append(roc_auc_score(y_test, y_pred))
        
        # Plot fold results
        plot_roc(y_test, y_probs_eegnet, 'EEGNet')
        plot_roc(y_test, y_probs_deepconv, 'DeepConvNet')
        plt.figure(figsize=(10,6))
        pd.DataFrame(results).plot(kind='box')
        plt.title('Model Comparison - Current Fold')
        plt.ylabel('ROC AUC')
        # plt.savefig(f'model_comparison_fold_{fold+1}.png')
        # plt.close()
        plt.show()
    
    # Final results
    print("\nFinal Results:")
    for model, scores in results.items():
        print(f"{model.upper():<20} Mean ROC-AUC: {np.mean(scores):.3f}  {np.std(scores):.3f}")
    
    return results

# Run evaluation
results = evaluate_models(X, y, groups)
```
